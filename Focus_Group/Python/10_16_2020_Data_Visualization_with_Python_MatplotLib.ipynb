#                      Data Visualization with Matplotlib

# to use Matplotlib plots on python we can use Matplotlib pyplot module.

```python
import matplotlib.pyplot as plt
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]

salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
```

```python
plt.plot(age, salary);
```

```python
plt.plot(age, salary)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");
```

```python
salary_2 = [45372, 48876, 53850, 57287, 63016, 65998, 70003, 70000, 71496, 75370, 83640]
```

```python
plt.plot(age, salary)
plt.plot(age, salary_2)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");
```

```python
plt.plot(age, salary, label="salary")
plt.plot(age, salary_2, label="salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age")
plt.legend();
```

```python
plt.subplot(2,1,1)
plt.plot(age, salary,"r")
plt.xlabel("age")
plt.ylabel("salary")
plt.subplot(2,1,2)
plt.plot(age, salary_2, "b")
plt.xlabel("age")
plt.ylabel("salary2")
plt.tight_layout();
```

```python
plt.subplot(1,2,1)
plt.plot(age, salary, "r")
plt.subplot(1,2,2)
plt.plot(age, salary_2, "b")
plt.tight_layout();
```

```python
plt.subplot(2,2,1)
plt.plot(age, salary, "r")
plt.subplot(2,2,2)
plt.plot(age, salary_2, "b")
plt.subplot(2,2,3)
plt.plot(age, salary, "g")
plt.subplot(2,2,4)
plt.plot(age, salary_2, "y")
plt.tight_layout();
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(age, salary, salary_2)), columns =['age', 'salary', 'salary_2'])
df.head()
```

```python
plt.plot(df["age"], df["salary"]);
```

```python
plt.plot(df.age, df.salary); 
```

# Object Oriented Method

```python
fig = plt.figure()
ax = fig.add_axes([0.6, 0.6, 0.9, 0.9])
```

```python
fig = plt.figure()
ax = fig.add_axes([0.0, 0.0, 0.8, 0.8])

ax.plot(age,salary, "r");
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("Salary by Age");
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.0, 0.0, 0.8, 0.8])

ax2 = fig.add_axes([0.0, 0.0, 0.5, 0.5])


ax1.plot(age,salary, "r")
ax1.set_xlabel("Age")
ax1.set_ylabel("Salary")
ax1.set_title("Salary by Age")

ax2.plot(age,salary_2, "b");
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary_2")
ax2.set_title("Salary_2 by Age");
```

```python
fig, ax = plt.subplots()
ax.plot(age, salary, "r")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("Salary by Age")
plt.tight_layout();
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(age,salary, "r")
ax[0,0].set_title("salary by age")
ax[0,1].plot(age, salary_2)
ax[0,1].set_title("salary_2 by age")
ax[1,0].plot(age,salary, "g")
ax[1,0].set_title("salary by age")
ax[1,1].plot(age, salary_2, "y")
ax[1,1].set_title("salary_2 by age")
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(8,3))

ax.plot(age, salary, label="salary_1")
ax.set_xlabel('age')
ax.plot(age, salary_2, label="salary2")
ax.set_title('Title')
ax.set_xlabel('age')
ax.set_ylabel('salary')
ax.legend();
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(6,3))

ax.plot(age, salary, label="salary_1")
ax.set_xlabel('age')
ax.plot(age, salary_2, label="salary2")
ax.set_title('Title')
ax.set_xlabel('age')
ax.set_ylabel('salary')
ax.legend();
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(age, salary, label="salary_1")
ax.set_xlabel('age')
ax.set_title('Title')
ax.set_ylabel('salary')
ax.legend();
```

```python
fig, ax = plt.subplots(figsize=(8,4))
# '-', '--', '-.', ':', ''
ax.plot(age, salary, "r", lw=5, linestyle = '--');
```

```python
fig, ax = plt.subplots(figsize=(8,4))
# '-', '--', '-.', ':', ''
ax.plot(age, salary, "r", lw=3, marker = "o", markersize = 25);
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(age, salary)
ax.set_xlim([30,40])
ax.set_ylim([45000,60000]);
```

```python
fig, ax = plt.subplots(figsize=(8,4))
# '-', '--', '-.', ':', ''
ax.plot(age, salary, "r", lw=3, ls = ' ', marker = "o", markersize = 10)
ax.axvline(x=35, color="green", ls="--")
ax.axhline(y=60000, color = "green", ls = "--");
```

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
```

```python
fig, ax = plt.subplots()
ax.bar(labels, men_means);
```

```python
fig, ax = plt.subplots()
ax.bar(labels, women_means, color="orange");
```

```python
fig, ax = plt.subplots()
ax.bar(labels, men_means)
ax.bar(labels, women_means, color="orange");
```

```python
import numpy as np
x = np.arange(len(labels))  # the label locations
x
```

```python
width = 0.35  # the width of the bars
```

```python
fig, ax = plt.subplots()

ax.bar(x - width/2, men_means, width=0.35, label='Men')
ax.bar(x + width/2, women_means, width=0.35, label='Women')
```

```python
fig, ax = plt.subplots()

ax.bar(x - width/2, men_means, width, label='Men')
ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
```
