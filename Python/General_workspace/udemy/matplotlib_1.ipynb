```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# bu komut sayesinde bundan sonra yapılmak istenen işlemler, çizdirilmek istenen grafikler tek komutta icra edilecek.
# Bu olmazsa her seferinde "plt.show()" komutunu kullanmalıyız.
```

```python
import numpy as np
x = np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

# Functional Method

```python
plt.plot(x,y);
```

```python
plt.plot(x,y, "r");
```

```python
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r")

plt.subplot(1,2,2)
plt.plot(y,x,"b")
```

# OO Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Set Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Larger Plot")

axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2) 

#axes.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")

plt.tight_layout() # sıkışık durumdaki chartları düzenler
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(8,2))

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title("Title")
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

```python
fig.savefig("my_picture.png", dpi=200) #hazırlanan grafiği istediğin formatta kaydededbilirsin.
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label="X Squared")
ax.plot(x,x**3, label="X Cubed")

ax.legend(loc=(0.01,0.8))
```

# Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color="orange", lw=3, alpha=0.5, ls="--", marker="o", markersize=10,
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="purple")
# also we can write RGB Hex Code here"
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color="purple", lw=2, ls="--")

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
