```python
import numpy as np
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.barplot(x="sex",y="total_bill",data=tips) # y sütunu mean değeri veriyor default olarak
```

```python
sns.barplot(x="sex",y="total_bill",data=tips, estimator=np.std) # burada std seçild y ekseni için
```

```python
sns.countplot(x="sex", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
sns.boxplot(x="total_bill", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue= "smoker")
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex")
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue="sex")
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue="sex",dodge=True) #dodge = split
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color="black")
```

```python
sns.factorplot(x="day", y="total_bill", data=tips, kind="bar") # kind parametresine ne yazarsan onu çizdirir.
```

```python
sns.factorplot(x="day", y="total_bill", data=tips, kind="violin")
```
