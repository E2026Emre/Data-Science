```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.countplot(x="sex", data=tips)
```

```python
sns.set_style("dark") # burada birçok seçenek var, dene ve gör {darkgrid, whitegrid, dark, white, ticks}
sns.countplot(x="sex", data=tips)
```

```python
sns.set_style("ticks") # burada birçok seçenek var, dene ve gör {darkgrid, whitegrid, dark, white, ticks}
sns.countplot(x="sex", data=tips)
sns.despine(left=True, bottom=True) # burada da seçenekler mevcut
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x="sex", data=tips)
```

```python
sns.set_context("poster", font_scale=1)
sns.countplot(x="sex", data=tips)
```

```python
sns.set_context("notebook", font_scale=1)
sns.countplot(x="sex", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex")
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", palette="coolwarm") # matplotlib'in sayfasından birçok renk seçeneği bulabilirsin
```
