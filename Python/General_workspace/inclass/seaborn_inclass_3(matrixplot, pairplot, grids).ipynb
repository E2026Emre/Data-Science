## spine removal

```python
plt.figure(figsize=(6,3))
plt.style.use('default')
sns.countplot(x="sex", data=tips)
sns.despine() # left, right, top and bottom parametreleri ile etraftaki izgilerin durumu deitirilebilir.
```
