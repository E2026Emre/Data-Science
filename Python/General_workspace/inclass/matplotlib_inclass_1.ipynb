```python
import matplotlib.pyplot as plt
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
```

```python
plt.plot(age, salary);
```

```python
plt.plot(age, salary)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");
```

```python
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
plt.plot(age, salary)
plt.plot(age, salary_2)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");
```

```python
plt.plot(age, salary, label = "salary")
plt.plot(age, salary_2, label = "salary_2")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age")
plt.legend();
```

```python
plt.subplot(1,2,1)
plt.plot(age, salary, "r")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age")

plt.subplot(1,2,2)
plt.plot(age, salary_2, "b")
plt.xlabel("age")
plt.ylabel("salary_2")
plt.title("Salary by Age")
plt.tight_layout();
```

```python
plt.subplot(2,2,1)
plt.plot(age, salary, "r")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age")

plt.subplot(2,2,2)
plt.plot(age, salary_2, "b")
plt.xlabel("age")
plt.ylabel("salary_2")
plt.title("Salary by Age")

plt.subplot(2,2,3)
plt.plot(age, salary, "g")
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age")

plt.subplot(2,2,4)
plt.plot(age, salary_2, "y")
plt.xlabel("age")
plt.ylabel("salary_2")
plt.title("Salary by Age")

plt.tight_layout();
```

```python
fig = plt.figure()
ax = fig.add_axes([0.0,0.0,0.8,0.8])

ax.plot(age,salary,"r")
ax.set_xlabel("age")
ax.set_ylabel("salary")
ax.set_title("Salary by Age");
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.0,0.0,0.8,0.8])
ax2 = fig.add_axes([0.1,0.4,0.3,0.3])

ax1.plot(age,salary,"r")
ax1.set_xlabel("age")
ax1.set_ylabel("salary")
ax1.set_title("Salary by Age")

ax2.plot(age,salary_2,"b")
ax2.set_xlabel("age")
ax2.set_ylabel("salary_2")
ax2.set_title("Salary_2 by Age");
```

```python
fig, ax = plt.subplots()
ax.plot(age,salary, "r")
ax.set_xlabel("age")
ax.set_ylabel("salary")
ax.set_title("Salary by Age")
plt.tight_layout();
```
