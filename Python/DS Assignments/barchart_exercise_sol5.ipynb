```python
import numpy as np
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("bar_data.csv")
```

```python
my_list =[]

for i in df['LanguagesWorkedWith']:
    a = i.split(";")
    for j in a:
        my_list.append(j)
```

```python
seri = pd.Series(my_list)
```

```python
seri.value_counts()
```

```python
df_g = pd.DataFrame(seri)
```

```python
df_g
```

```python
df_g['programing language'] = [i for i in range(1, len(my_list)+1)]
```

```python
df_g
```

```python
df_g_g = df_g.groupby(0).count()
```

```python
df_g_g
```

```python
df_g_g.drop(1, axis=1,inplace=True)
```

```python
df_g_g = df_g_g.reset_index()
```

```python
df_g_g
```

```python
df_g_g = df_g_g.sort_values(by = 'programing language')
```

```python
df_g_g
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,10))
ax.barh(df_g_g[0], df_g_g['programing language'])
ax.set_ylabel('')
ax.set_title('most popular languages')
ax.axvline(x=30000, color = 'r', ls = '--')
plt.tight_layout
```
