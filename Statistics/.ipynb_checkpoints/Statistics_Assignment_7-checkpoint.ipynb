### EXERCISE-1.Create a Standard Normal Distribution Table using Python scipy.stats.

### Solution-1

```python
columns = [.00, .01, .02, .03, .04, .05, .06, .07, .08, .09]
```

```python
rows = np.linspace(0,3.09,310)
```

```python
index = np.arange(0,3.1,0.1)
```

```python
values = np.array([(1-stats.norm.cdf(z,0,1)) for z in rows]) 
```

```python
distNorm = values.reshape(31,10)
```

```python
df_distNorm = round(pd.DataFrame(distNorm, columns=columns, index=index),4)
```

#### Solution-2

```python
def normalProbabilityDensity(x):
    constant = 1.0 / np.sqrt(2*np.pi)
    return (constant * np.exp((-x**2) / 2.0))
```

```python
std_norm_table = pd.DataFrame(data = [],
                             index = np.round(np.arange(0, 3.1, .1), 2),
                             columns = np.round(np.arange(0, 0.1, .01), 2))
```

```python
for index in std_norm_table.index:
    for column in std_norm_table.columns:
        z = np.round(index + column , 2)
        value, _ = quad(normalProbabilityDensity,np.NINF, z)
        std_norm_table.loc[index, column] = 1-value

std_norm_table.index = std_norm_table.index.astype(str)
std_norm_table.columns = [str(column).ljust(2,"0") for column in std_norm_table.columns]
```

```python
std_norm_table
```

```python
fraction=np.arange(0.00,0.1,0.01)
z_scores=np.arange(0, 3.1, 0.1)
zDist=dict()
for f in fraction:
    row=[]
    for z in z_scores:
        row.append(1-round(stats.norm.cdf(z+f),4))
    zDist.update({f:row})
pd.DataFrame(zDist)
```

### EXERCISE-2.The cycle time for trucks hauling concrete to a highway construction site is uniformly distributed over the interval 50 to 70 minutes. What is the probability that the cycle time exceeds 65 minutes  if it is known that the cycle time exceeds 55 minutes?

```python
(a, b) = (50, 20)
unifDist = stats.uniform(a, b)
# Event A: exceed 65
pA=1-unifDist.cdf(65)
# Event B: exceed 55
pB=1-unifDist.cdf(55)
# p(A | B ) = p(A n B)/p( B )
# p(A n B) = p( A )
pAB=pA/pB
pAB
```

```python
1-stats.uniform.cdf(65, 55, 15)
```

### EXERCISE-3.The width of bolts of fabric is normally distributed with mean 950 mm (millimeters) and standard deviation 10 mm.

#### What is the probability that a randomly chosen bolt has a width of between 947 and 958mm?

```python
stats.norm.cdf(958,950,10)-stats.norm.cdf(947,950,10)
```

#### What is the appropriate value for C such that a randomly chosen bolt has a width less than C with probability .8531?

```python
# z score for (.8531) = 1.05
mu =950
sigma = 10
z = 1.05
x = (z * sigma) + mu
x
```

```python
stats.norm.ppf(0.8531,950,10)
```

### EXERCISE-4.The school board administered an IQ test to 20 randomly selected teachers. They found that the average IQ score was 114 with a standard deviation of 10. Assume that the cumulative probability is 0.90. What population mean would have produced this sample result?

### solution-1

```python
# t value for cum.prob = .90 and df = 19 => 1.328
x = 114
n = 20
s = 10
t = 1.328
mu = x - (t * s/math.sqrt(n))
mu
```

```python
stats.t.cdf(1.328, df=19)
```
